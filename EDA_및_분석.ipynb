### 업종별 폐업확률

```python
# 업종별
result = df.groupby('업종').agg(
    평균폐업확률=('폐업확률', 'mean'),
    가맹점수=('업종', 'count')
).sort_values('평균폐업확률', ascending=False).head(20)

result[result['가맹점수'] >= 10].sort_values('평균폐업확률', ascending=False).head(50)
```

```python
# 지역별
import re

# '가맹점주소' 간소화
df['가맹점주소'] = df['가맹점주소'].str.replace(r'\d+(?=가길|\w*길)', '', regex=True)
df['가맹점주소'] = df['가맹점주소'].str.replace(r'\s+\d+(-\d+)?$', '', regex=True)
df['가맹점주소'] = df['가맹점주소'].str.replace(r'\s+[\d\-\.]+$', '', regex=True)
df['가맹점주소'] = df['가맹점주소'].str.replace(r'\d+.*$', '', regex=True).str.strip()
print("주소 전처리 완료")

# 주소 정규화
df['가맹점주소'] = df['가맹점주소'].str.replace('서울특별시', '서울', regex=False)
df['가맹점주소'] = df['가맹점주소'].str.replace(r'\s+', ' ', regex=True).str.strip()
for _ in range(3):
    df['가맹점주소'] = df['가맹점주소'].str.replace(r'(로|길|동|가|나|다|라|마|번|지하)$', '', regex=True).str.strip()

df[df['폐업여부'] == 1].groupby('가맹점주소').size().sort_values(ascending=False).head(10)
```

### 폐업하지 않은 가맹점의 평균 폐업확률 (가맹점수 100개 이상)

```python
result = df[df['폐업여부'] == 0].groupby('가맹점주소').agg(
    평균폐업확률=('폐업확률', 'mean'),
    가맹점수=('폐업확률', 'count')
)
result[result['가맹점수'] >= 100].sort_values('평균폐업확률', ascending=False).head(50)
```

- 왕십리가 가장 폐업한 가맹점이 많다.

# 의사결정나무를 통한 임계치 설정

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# --- 2. 분석할 주요 변수 목록 정의 ---
features_to_analyze = [
    '동일업종내매출순위비율', '동일상권내매출순위비율',
    '재방문고객비중', '신규고객비중',
    '거주이용고객비율', '직장이용고객비율', '유동인구이용고객비율',
    '배달매출금액비율', '동일업종매출금액비율'
]

results = []

# --- 3. 각 변수별로 반복하면서 최적의 임계점 찾기 및 시각화 ---
print("\n각 변수별 위험 임계점 분석 및 시각화를 시작합니다...")
for feature in features_to_analyze:
    temp_df = df[[feature, '폐업여부']].dropna()

    if len(temp_df) < 100:
        continue

    X = temp_df[[feature]]
    y = temp_df['폐업여부']

    tree_model = DecisionTreeClassifier(max_depth=1, random_state=42)
    tree_model.fit(X, y)

    threshold = tree_model.tree_.threshold[0]

    if threshold != -2:
        high_risk_group = temp_df[temp_df[feature] <= threshold]
        low_risk_group = temp_df[temp_df[feature] > threshold]

        if len(high_risk_group) == 0 or len(low_risk_group) == 0:
            continue

        high_risk_closure_rate = high_risk_group['폐업여부'].mean()
        low_risk_closure_rate = low_risk_group['폐업여부'].mean()

        impact = 0
        if low_risk_closure_rate > 0:
            impact = high_risk_closure_rate / low_risk_closure_rate
        else:
            impact = float('inf')

        results.append({
            '변수명': feature,
            '위험 임계점': f"{threshold:.2f} 이하",
            '위험 그룹 폐업률': f"{high_risk_closure_rate:.2%}",
            '안전 그룹 폐업률': f"{low_risk_closure_rate:.2%}",
            '영향력 (배수)': f"{impact:.2f} 배"
        })

        # --- 의사결정나무 시각화 코드 추가 ---
        plt.figure(figsize=(12, 7))
        plot_tree(tree_model,
                  feature_names=[feature],
                  class_names=['생존', '폐업'],
                  filled=True,
                  proportion=True,
                  rounded=True,
                  fontsize=12)
        plt.title(f"'{feature}' 변수의 폐업/생존 분기점", fontsize=16)
        plt.show()


# --- 4. 분석 결과 정리 및 출력 ---
print("분석이 완료되었습니다.")
if results:
    results_df = pd.DataFrame(results)
    results_df['영향력 (배수)_float'] = results_df['영향력 (배수)'].apply(lambda x: float(x.replace(' 배', '')))
    results_df_sorted = results_df.sort_values(by='영향력 (배수)_float', ascending=False).drop(columns=['영향력 (배수)_float'])

    print("\n## 각 변수별 폐업 위험 임계점 및 영향력 분석 결과 ##")
    print(results_df_sorted.to_string())
else:
    print("\n분석 결과가 없습니다.")
```

### 모델링 전 전처리

```python
# -999999.9 값을 실제 결측치(NaN)으로 변환
df.replace(-999999.9, np.nan, inplace=True)

# datetime 으로 전처리
df['개설일'] = pd.to_datetime(df['개설일'], format='%Y%m%d', errors='coerce')
df['폐업일'] = pd.to_datetime(df['폐업일'], format = '%Y%m%d', errors='coerce')
df['기준년월'] = pd.to_datetime(df['기준년월'], format='%Y%m', errors='coerce')

# 전체 데이터에서 가장 마지막관측시점 계산
end_of_study_date = df['기준년월'].max()

#운영기간 계산
df['duration'] = np.where(df['폐업일'].notna(),
                          (df['폐업일'] - df['개설일']).dt.days,
                          (end_of_study_date - df['개설일']).dt.days)
df['event'] = df['폐업여부']
```

```python
from lifelines import CoxPHFitter
from sklearn.preprocessing import StandardScaler

# 가맹점운영개월수 구간을 숫자로 변환
def map_duration(s):
  if pd.isna(s): return np.nan
  if '50% - 75%' in s: return 30
  if '75-90%' in s: return 40
  if '90%초과' in s: return 50
  return 15

# 0일 보다 짧은 운영기간이나 날짜 오류가 있는 데이터 제거
df = df[df['duration'] > 0].copy()

#분석에 필요한 변수 목록
covariates = [
    '재방문고객비중', '신규고객비중', '배달매출금액비율',
    '동일업종매출금액비율', '동일업종내매출순위비율', '동일상권내매출순위비율',
    '거주이용고객비율'
]

# 결측치 제거
analysis_df = df[['duration', 'event'] + covariates].dropna()

# 데이터 스케일링: StandardScaler를 사용하여 변수들을 표준화
scaler = StandardScaler()
analysis_df[covariates] = scaler.fit_transform(analysis_df[covariates])

# 콕스 비례위험 모델 학습
cph = CoxPHFitter(penalizer=0.1) # Ridge 페널티를 추가하여 모델 안정성 향상
cph.fit(analysis_df, duration_col='duration', event_col='event', show_progress=True)

#결과 해석
print(cph.summary[['exp(coef)','p']])

# Hazard Ratio 시각화
plt.figure(figsize = (10,6))
cph.plot()
plt.title('폐업 위험도 영향요인', fontsize=16)
plt.ylabel('상대적 위험도')
plt.show()

hr_summary = cph.summary.sort_values(by='exp(coef)')

# Hazard Ratio가 1보다 작으면 생존에 긍정적
positive_factors = hr_summary[hr_summary['exp(coef)'] < 1]
if not positive_factors.empty:
    best_factor = positive_factors.iloc[0]
    hr_best = best_factor['exp(coef)']
    reduction = (1 - hr_best) * 100
    print(f"가장 강력한 생존 요인: {best_factor.name}")
    print(f"-> 해석: 이 지표가 1단위 증가할 때마다, 폐업 위험이 {reduction:.1f}%씩 감소.")

# Hazard Ratio가 1보다 크면 생존에 부정적
negative_factors = hr_summary[hr_summary['exp(coef)'] > 1]
if not negative_factors.empty:
    worst_factor = negative_factors.iloc[-1]
    hr_worst = worst_factor['exp(coef)']
    increase = (hr_worst - 1) * 100
    print(f"\n가장 강력한 위험 요인: {worst_factor.name}")
    print(f"-> 해석: 이 지표가 1단위 증가할 때마다, 폐업 위험이 {increase:.1f}%씩 증가.")
```

# 폐업 전 매출 패턴 파악

```python
import warnings

warnings.filterwarnings('ignore')

# 분석 기간 (폐업 기준 몇 개월 전부터 몇 개월 전까지)
months_before_closure_start = -6 # 폐업 6개월 전부터 분석
months_before_closure_end = -1   # 폐업 1개월 전까지 분석

numerical_cols = ['배달매출금액비율', '동일업종매출금액비율', '재방문고객비중', '신규고객비중',
                   '거주이용고객비율', '직장이용고객비율', '유동인구이용고객비율'] # 수치형 변수
categorical_cols = ['매출금액구간', '매출건수구간', '유니크고객수구간', '객단가구간'] # 범주형 변수

# 폐업 그룹과 생존 그룹 분리
df_closed = df[(df['폐업여부'] == 1) & (df['폐업일'].notna())].copy()
df_surviving = df[df['폐업여부'] == 0].copy()

# 폐업 그룹: '폐업까지 남은 개월 수' 계산 (정확한 월 차이 계산)
df_closed['폐업일_period'] = df_closed['폐업일'].dt.to_period('M')
df_closed['기준년월_period'] = df_closed['기준년월'].dt.to_period('M')

df_closed['폐업일_period'] = df_closed['폐업일'].dt.to_period('M')
df_closed['기준년월_period'] = df_closed['기준년월'].dt.to_period('M')
df_closed['폐업까지_개월수'] = (df_closed['기준년월_period'] - df_closed['폐업일_period']).apply(lambda x: x.n if pd.notna(x) else np.nan)
df_closed.dropna(subset=['폐업까지_개월수'], inplace=True)
df_closed['폐업까지_개월수'] = df_closed['폐업까지_개월수'].astype(int)

# 분석 기간 데이터 필터링 (폐업 그룹)
df_closed_period = df_closed[
  (df_closed['폐업까지_개월수'] >= months_before_closure_start) &
  (df_closed['폐업까지_개월수'] <= months_before_closure_end)
].copy()

# 생존 그룹: 비교를 위한 '상대 개월 수' 계산
df_surviving['month_rank_desc'] = df_surviving.groupby('가맹점구분번호')['기준년월'].rank(method='first', ascending=False)
n_months = abs(months_before_closure_start - months_before_closure_end) + 1
df_surviving_period = df_surviving[df_surviving['month_rank_desc'] <= n_months].copy()
df_surviving_period['상대_개월수'] = (df_surviving_period['month_rank_desc'] * -1).astype(int) + 1
df_surviving_period = df_surviving_period[
      (df_surviving_period['상대_개월수'] >= months_before_closure_start) &
      (df_surviving_period['상대_개월수'] <= months_before_closure_end)
]
```

```python
# 그룹별 월 평균 계산
closed_trends_num = df_closed_period.groupby('폐업까지_개월수')[numerical_cols].mean()
surviving_trends_num = df_surviving_period.groupby('상대_개월수')[numerical_cols].mean()

# 수치형 변수 추세 시각화
num_plots = len(numerical_cols)
n_cols_num = 3 # 한 줄에 그래프 3개씩
n_rows_num = (num_plots + n_cols_num - 1) // n_cols_num
plt.figure(figsize=(15, 5 * n_rows_num))

for i, col in enumerate(numerical_cols):
    ax = plt.subplot(n_rows_num, n_cols_num, i + 1)
    ax.plot(closed_trends_num.index, closed_trends_num[col], marker='o', linestyle='-', label='폐업 그룹')
    ax.plot(surviving_trends_num.index, surviving_trends_num[col], marker='x', linestyle='--', label='생존 그룹')
    ax.set_title(f'{col} 변화 추세')
    ax.set_xlabel('폐업 기준 상대 개월수 (0=폐업월)')
    ax.set_ylabel('평균값')
    ax.legend()
    ax.grid(True)

plt.suptitle('폐업 vs 생존 그룹 수치형 변수 월별 변화 추세', size=16, y=1.02)
plt.tight_layout()
plt.savefig('numerical_trends_comparison.png') # 그래프 파일 저장
print("수치형 변수 추세 그래프 저장 완료: numerical_trends_comparison.png")
```
